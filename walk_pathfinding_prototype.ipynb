```python
import networkx as nx
import matplotlib.pyplot as plt
from kp_registry import Registry
import os
os.environ["KP_TRAPI_VERSION"] = "1.4"
```

```python
registry = Registry()
kps = await registry.retrieve_kps()
print(len(kps.items()))
```

```python
G = nx.DiGraph()
for kp, details in kps.items():
    for operation in details["operations"]:
        G.add_edge(operation["subject_category"], operation["object_category"], predicate=operation["predicate"])
```

```python
plt.figure(1, figsize = (30,30))
nx.draw_random(G, with_labels=True)
plt.show()
```
